```python
import pandas as pd
import math
import time
start = time.clock()

df = pd.read_excel('zhangwu.xlsx').drop(columns = ['Unnamed: 0'])

df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
df['month'] = df['Date'].apply(lambda x : x.strftime('%m'))
df = df.sort_values(by=['month','Date'])

df['x'] = df.apply(lambda x: math.sin(x['Dir']) * x['Spd'], axis=1)
df['y'] = df.apply(lambda x: math.cos(x['Dir']) * x['Spd'], axis=1)
df['wind_energy'] = df.apply(lambda x: 0.5 * 1.225 * math.pow(x['Spd'],2), axis=1)

dfs = df.describe(include='all')

df['Date'] = df['Date'].apply(lambda x : x.strftime('%Y-%m-%d')).apply(str)

df1 = df.loc[df['month'] == '01'].drop(columns = ['month'])
df2 = df.loc[df['month'] == '02'].drop(columns = ['month'])
df3 = df.loc[df['month'] == '03'].drop(columns = ['month'])
df4 = df.loc[df['month'] == '04'].drop(columns = ['month'])
df5 = df.loc[df['month'] == '05'].drop(columns = ['month'])
df6 = df.loc[df['month'] == '06'].drop(columns = ['month'])
df7 = df.loc[df['month'] == '07'].drop(columns = ['month'])
df8 = df.loc[df['month'] == '08'].drop(columns = ['month'])
df9 = df.loc[df['month'] == '09'].drop(columns = ['month'])
df10 = df.loc[df['month'] == '10'].drop(columns = ['month'])
df11 = df.loc[df['month'] == '11'].drop(columns = ['month'])
df12 = df.loc[df['month'] == '12'].drop(columns = ['month'])
dfmerge1 = pd.concat([df3,df4,df5,df6,df7])
dfmerge2 = pd.concat([df1,df2,df8,df9,df10,df11,df12])

print ("processing complete")
```

```python
with pd.ExcelWriter('zhangwu_wind_output_1.xlsx') as writer: 
    dfs.to_excel(writer, sheet_name='summary')
    df.drop(columns = ['month']).to_excel(writer, sheet_name='00',index=None)
    df1.to_excel(writer, sheet_name='01', index=None)
    df2.to_excel(writer, sheet_name='02', index=None)
    df3.to_excel(writer, sheet_name='03', index=None)
    df4.to_excel(writer, sheet_name='04', index=None)
    df5.to_excel(writer, sheet_name='05', index=None)
    df6.to_excel(writer, sheet_name='06', index=None)
    df7.to_excel(writer, sheet_name='07', index=None)
    df8.to_excel(writer, sheet_name='08', index=None)
    df9.to_excel(writer, sheet_name='09', index=None)
    df10.to_excel(writer, sheet_name='10', index=None)
    df11.to_excel(writer, sheet_name='11', index=None)
    df12.to_excel(writer, sheet_name='12', index=None)
    dfmerge1.to_excel(writer, sheet_name='3-7', index=None)
    dfmerge2.to_excel(writer, sheet_name='8-2', index=None)
    
print ("write over")
```

```python
end = time.clock()
print('Running time: %s Seconds'%(end-start))
```
